```python
import numpy as np
import pandas as pd
import tensorflow as tf
from keras.utils import to_categorical
from keras.models import Sequential
from keras import layers
from keras import regularizers
```

```python
data = pd.read_csv("ionosphere_data.csv")
```

```python
data.head()
```

```python
data = pd.get_dummies(data,columns = ["label"])
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
Input = data.iloc[:,:-2]
Output = data.iloc[:,-2:]
X_train = Input.iloc[:210,:]
Y_train = Output.iloc[:210,:]
X_test = Input.iloc[210:,:]
Y_test = Output.iloc[210:,:]
```

```python
def normalize(x):
  mean_x = x.mean(axis = 0)
  std_x = x.std(axis =0)
  x-=mean_x
  x/=std_x

normalize(X_train)
normalize(X_test)
```

```python
Input.head()
```

```python
Output.head()
```

```python
model = Sequential()
model.add(layers.Dense(16,activation = "relu", kernel_regularizer= regularizers.l2(0.001), input_shape = (X_train.shape[1],)))
model.add(layers.Dense(8, kernel_regularizer=regularizers.l2(0.001),activation="relu"))
model.add(layers.Dense(2,activation= "sigmoid"))
```

```python
model.compile(optimizer = "rmsprop",loss = "binary_crossentropy",metrics = ["accuracy"])
```

```python
model.fit(Input,Output,epochs= 15,batch_size = 10)
```

```python
model.evaluate(X_train,Y_train)
```

```python
model.predict(X_test)
```
